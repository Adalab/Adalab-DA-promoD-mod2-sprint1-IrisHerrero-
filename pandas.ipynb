- Hay algunas canciones cuyo título está en mayúsulas. Crea una función para que todos los títulos estén en minúscula. Haced lo mismo para las columnas de "track" y "album".

```python
def float_sin_comas(x):
    try:
        x = float(x.replace(",", "."))
        return x
    except:
        return x
```

```python
lista_floats = ['danceability', 'energy', 'loudness', 'speechiness', 'key', 'acousticness', 'instrumentalness']
```

```python
for col in lista_floats:
    df[col]=df[col].apply(float_sin_comas)
```

```python
def titulos_minuscula(x):

    try:
       x = x.lower()
       return x
    except:
        return np.nan
```

```python
df.sample()
```

```python
lista_lower = ['title', 'track', 'album']
```

```python
for col in lista_lower:
    df[col] = df[col].apply(titulos_minuscula)
```

- Algunos de los artistas tienen símbolos raros, en concreto el símbolo $. Cread una función que nos elimine este símbolo de los nombres de los artistas.

```python
patron1= '\$'
patron2 = 's'
```

```python
def quitar_simbolo(antiguo, nuevo, columna):

    try:
        simbolo= re.sub(antiguo, nuevo, columna) #método .sub de regex para reflejar, sustituir, y lugar donde sustituir
        return simbolo
    except:
        return np.nan
```

```python
df['artist']= df.apply(lambda dataframe: quitar_simbolo(patron1, patron2, dataframe['artist']), axis=1)
```

- Se quiere estudiar si existe una relación entre la cantidad de visitas que reciben los videos de música de un artista en una plataforma de streaming y la cantidad de likes que estos videos obtienen. Por lo tanto, querremos contestar a la siguiente pregunta: ¿Son aquellos artistas que tienen más visitas los que más likes tienen? Para solucionar este ejercicio deberéis:

   -Agrupar por artista y calcular la media de visitas y de likes.

```python
artista_vistas = df.groupby(['artist'])['views'].mean().reset_index().sort_values(by='views', ascending=False)
artista_vistas.head() #agrupamos por artistas y que nos muestre los más vistos de media
```

```python
artista_likes = df.groupby(['artist'])['likes'].mean().reset_index().sort_values(by='likes', ascending=False)
artista_likes.head() ##agrupamos por artistas y que nos muestre el que tiene más likes
```

  -Quedaros con los 10 artistas que más visitas han tenido.

```python
top_10_visitas = artista_vistas.head(10)
top_10_visitas 
```

-Quedaros con los 10 artistas que más likes han tenido.

```python
top_10_likes = artista_likes.head(10)
top_10_likes
```

-Haced un subplot, donde en uno tengamos los 10 artistas con más likes y en el otro los 10 artistas con mas views.

```python
fig, axes = plt.subplots(1,2, figsize =(15,5))

sns.barplot(x='views', y= 'artist', data = top_10_visitas, ax= axes[0])
sns.barplot(x='likes', y= 'artist', data = top_10_likes, ax= axes[1])

plt.tight_layout()
plt.show()
```
